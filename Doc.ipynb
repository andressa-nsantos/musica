# **Passos realizados no projeto**

## **1) Adaptação do código para a obtenção das músicas**

1) Baixei o chromedriver e suas extensões ao invés do geckodriver e o navegador Firefox; isso trouxe algumas mudanças nas bibliotecas e funções importadas no início do código;
2) Adaptei as partes do código de "api.py" em "data" que possuiam estrutura semelhante a essa: [By.CSS_SELECTOR, "div.cifra_cnt a"] para atender as necessidades do site buscado (cifraclub.com);
3) Inclui a captura das variáveis 'tom', 'capotraste', 'compositor'e substitui 'letra' por 'letra_e_cifra';.

Obs.: A variável 'capotraste' é relevante pois certas músicas estão em um tom, porém sua cifra não corresponde a ele, e isso acontece porque há o uso de capotraste na canção. Assim, tendo conhecimento dessa variável, seremos capazes de adaptar as notas da cifra ao tom original e fazer uma análise mais completa.

## **2 - Processo de obtenção dos dados**

1) Mudei o banco de dados original para conter apenas ID, nome da música, nome do artista e o ano; esse novo banco chama-se 'raw_data.csv' e está na pasta 'data';
2) Cada música foi armazenada em um arquivo com extensão .json na pasta "processed";
3) Criei a pasta "processed - per year" para armazenar os mesmos arquivos da pasta "processed", mas ordenadas por ano; além disso, há o arquivo "musics found per year.xlsx" nesta pasta, que indica o total de músicas com cifras disponíveis dentre as aproximadamente 100 mais ouvidas de cada ano.

Obs.: o arquivo 'dados_reinaldo.csv' contém os dados finalizados do trabalho de Reinaldo, que adaptei para obter o novo banco 'raw_data.csv'

### **2.1 - Controle das músicas

1) Renomeação das músicas no arquivo "raw_data.csv", que não reconheceu as acentuações e substituiu pelo símbolo "�". Foi alterada música por música, porém não as iniciais

2) Criação da planilha "dados músicas.xlsx" com colunas representando ID da música, nome da música, nome do interprete e ano do emplacamento. Além disso, foi inserida uma coluna com cores.
2.1 - Simbologia da coluna colorida:
- verde: a música foi inserida corretamente pelo bot
- verde com preenchimento: o bot não captou a música, porém ela possui cifra disponível (do cifraclub - site referência) no link inserido e foi baixada
- vermelho: a música foi inserida, porém errada pelo bot (outra música, que não a pedida)
- vermelho com preenchimento: a música foi inserida, porém errada pelo bot (outra música, que não a pedida) e, portanto, foi eliminada. Porém, ela possui cifra disponível (não do cifraclub - site referência) no link inserido
- branco: não foi encontrada a cifra em sites
- branco com preenchimento: o bot não captou a música, porém ela possui cifra disponível (não do cifraclub - site referência) no link inserido

3) Foram limitados os seguintes sites para obtenção de cifras: "Cifra Club", "Ultimate Guitar", "Cifras", e "Banana Cifras;
3.1 - Padrão
- o bot original pesquisou as cifras no site "Cifra Club", por sua completude de informações. As músicas com cifras disponíveis em outros sites foram abrangidas por seus links posteriormente
- foi priorizada a cifra em relação a outros aspectos, como os compositores. Sendo assim, há casos em que existe a informação sobre a composição, mas não sobre as cifras. Neste caso, a música não foi colhida.

obs.: O ano de 1991 não constava no banco de dados
