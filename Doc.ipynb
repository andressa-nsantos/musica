# **Passos realizados no projeto**

## **1) Adaptação do código para a obtenção das músicas**

1) Baixei o chromedriver e suas extensões ao invés do geckodriver e o navegador Firefox; isso trouxe algumas mudanças nas bibliotecas e funções importadas no início do código;
2) Adaptei as partes do código de "api.py" em "data" que possuiam estrutura semelhante a essa: [By.CSS_SELECTOR, "div.cifra_cnt a"] para atender as necessidades do site buscado (cifraclub.com);
3) Inclui a captura das variáveis 'tom', 'capotraste', 'compositor'e substitui 'letra' por 'letra_e_cifra';.

Obs.: A variável 'capotraste' é relevante pois certas músicas estão em um tom cuja cifra não corresponde. Isso acontece porque há o uso de capotraste na canção. Assim, tendo conhecimento dessa variável, seremos capazes de adaptar as notas da cifra ao tom original e fazer uma análise mais correta.

## **2 - Processo de obtenção dos dados**

1) Mudei o banco de dados original para conter apenas ID, nome da música, nome do artista e o ano; esse novo banco chama-se 'raw_data.csv' e está na pasta 'data';
2) Cada música foi armazenada em um arquivo com extensão .json na pasta "processed";
3) Criei a pasta "processed - per year" para armazenar os mesmos arquivos da pasta "processed", mas ordenadas por ano; além disso, há o arquivo "musics found per year.xlsx" nesta pasta, que indica o total de músicas com cifras disponíveis dentre as aproximadamente 100 mais ouvidas de cada ano.

Obs. 1: o arquivo 'dados_reinaldo.csv' contém os dados finalizados do trabalho de Reinaldo, que adaptei para obter o novo banco 'raw_data.csv'
Obs. 2: Não dei prosseguimento à continuação da pasta "processed - per year", pois não é prioridade.

### 2.1 - Controle das músicas

1) Foi realizada a renomeação das músicas no arquivo "raw_data.csv", por não serem reconhecidas no arquivo as acentuações, aue foram substituídas inicialmente pelo símbolo "�". Foram alteradas música por música, exceto as iniciais.

2) Criação da planilha "dados músicas.xlsx" com colunas representando ID da música, nome da música, nome do intérprete e ano do emplacamento. Além disso, foi inserida uma coluna com cores.
2.1 - Simbologia da coluna colorida:
- verde: a música foi inserida corretamente pelo bot ou manualmente após pelo processo explicado no item 4)
- vermelho: a música foi inserida, porém errada pelo bot
- branco: não foi encontrada a cifra em sites
- branco/vermeho com preenchimento: o bot não captou a música, porém ela possui cifra disponível (não do cifraclub - site referência) no link inserido

### 2.2 - Controle das fontes

Foram limitados os seguintes sites para obtenção de cifras: "Cifra Club", "Ultimate Guitar", "Cifras", e "Banana Cifras;

- o bot original pesquisou as cifras no site "Cifra Club", por sua completude de informações. As músicas com cifras disponíveis em outros sites foram abrangidas por seus links posteriormente;
- foi priorizada a cifra em relação a outros aspectos, como os compositores. Sendo assim, há casos em que existe a informação sobre a composição, mas não sobre as cifras. Neste caso, a música não foi colhida.

Obs. 1: O site Cifra Club mudou seu front-end durante o download dos dados. Poe isso, a primeira parte do código da API está comentada (era a forma de coletar os dados na antiga interface);

### 2.3 - Controle manual

4) Muitas músicas foram inseridas erradas pelo bot. Por isso, corrigi manualmente linha a linha se o nome da música e do autor estavam corretos. Para realizar as correções, criei os arquivos "api_revisao.py" e "review_data.csv", análogas a "api.py" e "raw_data". Assim, há como mnsurar o número de erros cometidos no webscrapping (consulltando o último arquivo) e foi a maneira como reinseri as músicas existentes que não haviam sido capturadas.
